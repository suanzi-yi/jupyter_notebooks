# 多表拼接

## 横向拼接merge()

```python
df1=pd.DataFrame([[1,'小张',100,650],[2,'小王',101,600]],columns=['名次','姓名','学号','成绩'])
df2=pd.DataFrame([[100,'一班'],[101,'二班']],columns=['学号','班级'])
```

```python
pd.merge(df1,df2)#以公共列横向一对一拼接
```

```python
df3=pd.DataFrame([['小张',100,650],['小王',101,600],['小李',102,578]],columns=['姓名','学号','f_成绩'])
df4=pd.DataFrame([[100,586],[100,602],[101,691],[101,702],[102,645],[102,676]],columns=['学号','e_成绩'])
```

```python
pd.merge(df3,df4,on='学号')#公共列多对一拼接
```

```python
df5=pd.DataFrame([['小张',100,650],['小张',100,610],['小王',101,600],['小李',102,578],['小李',102,542]],columns=['姓名','学号','f_成绩'])
```

```python
pd.merge(df5,df4)#公共列多对对拼接
```

## 连接键的类型merge(on='')

```python
pd.merge(df1,df2)#默认以公共列作为连接键
```

```python
pd.merge(df1,df2,on='学号')#on指定连接键
```

```python
df2.insert(0,'姓名',['小张','小王'])
```

```python
pd.merge(df1,df2,on=['姓名','学号'])#多个公共列作为连接键
```

```python
df6=df1.rename(columns={'学号':'编号'})
df7=df2.drop(['姓名'],axis=1)
```

```python
pd.merge(df6,df7,left_on='编号',right_on='学号')#分别指定左右连接键
```

```python
df8=df6.set_index('编号')
df9=df7.set_index('学号')
```

```python
pd.merge(df8,df9,left_index=True,right_index=True)#把索引当做连接键
```

```python
pd.merge(df8,df7,left_index=True,right_on='学号')#索引和列连接
```

## 连接类型merge(how=' ')

```python
df10=pd.DataFrame([[1,'小张',100,650],[2,'小王',101,600],[3,'小赵',103,550]],columns=['名次','姓名','学号','成绩'])
df11=pd.DataFrame([['小张',100,'一班'],['小王',101,'一班'],['小钱',104,'三班']],columns=['姓名','学号','班级'])
```

```python
pd.merge(df10,df11,on='学号',how='inner')#内连接
```

```python
pd.merge(df10,df11,on='学号',how='left')#左连接
```

```python
pd.merge(df10,df11,on='学号',how='right')#右连接
```

```python
pd.merge(df10,df11,on='学号',how='outer')#外连接
```

### 重复列名处理merge(suffixes=[])

```python
pd.merge(df10,df11,on='学号',how='inner',suffixes=['_L','_R'])#自定义重复列名
```
